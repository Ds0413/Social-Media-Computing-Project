## Tweet-preprocessor Cleaning

```python
p.set_options(p.OPT.URL, p.OPT.MENTION, p.OPT.RESERVED, p.OPT.EMOJI)

def clean_tweet(text):
    return p.clean(text)

df_copy['clean_text'] = df_copy['text'].astype(str).apply(clean_tweet)
df_copy[["text", "clean_text"]].head()
```

## Loading Slang Dict

```python
df_slang = pd.read_csv(d_path + "slang/slang.csv")

slang_dict = dict(zip(df_slang.acronym.str.lower(),
                      df_slang.expansion.str.lower()))
```

## Slang Expansion

```python
def expand_slang(text):
    text = contractions.fix(text)
    pattern = re.compile(r"\b(" + "|".join(map(re.escape, slang_dict)) + r")\b")
    return pattern.sub(lambda m: slang_dict[m.group(0)], text)

df_copy['clean_expand'] = df_copy['clean_text'].apply(expand_slang)
df_copy[["text", "clean_text", "clean_expand"]].head()
```

## Tokenize, remove stopwords, stem, POS

```python
tknzr      = TweetTokenizer(strip_handles=True, reduce_len=True)
stop_words = set(stopwords.words('english'))
stemmer    = PorterStemmer()

def tokenize_process(text):
    words = tknzr.tokenize(text.lower())
    words = [w for w in words if w not in stop_words]
    stems = [stemmer.stem(w) for w in words]
    tags  = nltk.pos_tag(stems)
    return stems, tags

df_copy[['tokens','pos_tags']] = df_copy['clean_expand'].apply(
    lambda x: pd.Series(tokenize_process(x))
)

df_copy[["text", "clean_text", "clean_expand", "tokens", "pos_tags"]].head()
```

```python
from transformers import pipeline
import pandas as pd
from sklearn.metrics import classification_report, accuracy_score

# Using a sentiment analysis model from the Hugging Face Transformers library
sentiment_pipeline = pipeline("sentiment-analysis")

# Function to get sentiment prediction from the new transformer model
def predict_sentiment_transformer(tokens):
    # Join tokens into a single string
    text = " ".join(tokens)

    if pd.isna(text) or text == "":
        return None, None
    try:
        result = sentiment_pipeline(text)[0]
        # The pipeline returns a dictionary with 'label' and 'score'
        return result['label'], result['score']
    except Exception as e:
        print(f"Error processing text: {text[:50]}... Error: {e}")
        return None, None

# Assuming df_copy is already defined from the previous code
df_copy[['transformer_sentiment', 'transformer_confidence']] = df_copy['tokens'].apply(
    lambda x: pd.Series(predict_sentiment_transformer(x))
)

# Print the DataFrame with new Transformer predictions
print(df_copy[['airline_sentiment', 'clean_expand', 'flair_sentiment_mapped', 'transformer_sentiment', 'transformer_confidence']].head())

print("\nUnique sentiment labels from the new transformer:")
print(df_copy['transformer_sentiment'].unique())

transformer_label_mapping = {
    'POSITIVE': 'positive',
    'NEGATIVE': 'negative',
    'NEUTRAL': 'neutral'
}

# Apply the mapping. Handle potential NaNs before mapping.
df_copy['transformer_sentiment_mapped'] = df_copy['transformer_sentiment'].fillna('Unknown').map(transformer_label_mapping).fillna('Unknown')

# Filter for evaluation - ensure both original and mapped transformer sentiments are valid
eval_df_transformer = df_copy[df_copy['transformer_sentiment_mapped'] != 'Unknown'].copy()

# Filter eval_df_transformer to include only labels present in original sentiment for a fair comparison
valid_original_sentiments = df_copy['airline_sentiment'].unique().tolist()  # ['neutral', 'positive', 'negative']
eval_df_transformer = eval_df_transformer[eval_df_transformer['airline_sentiment'].isin(valid_original_sentiments)]

print("\nClassification Report for New Transformer Model Predictions:")
unique_eval_labels = sorted(list(set(eval_df_transformer['airline_sentiment'].unique()).union(set(eval_df_transformer['transformer_sentiment_mapped'].unique()))))

print(classification_report(
    eval_df_transformer['airline_sentiment'],
    eval_df_transformer['transformer_sentiment_mapped'],
    labels=unique_eval_labels,
    target_names=unique_eval_labels
))

print("\nAccuracy Score for New Transformer Model Predictions:")
print(accuracy_score(eval_df_transformer['airline_sentiment'], eval_df_transformer['transformer_sentiment_mapped']))
```

## Load GloVe embeddings & Build Embedding Matrix

```python
embedding_dim = 100
emb_index = {}
glove_dir = '/content/drive/MyDrive/Social Media Computing/glove_data/'
with open(glove_dir+'glove.6B.100d.txt','r',encoding='utf8') as f:
    for line in f:
        vals = line.split()
        emb_index[vals[0]] = np.asarray(vals[1:], dtype='float32')
```

## Tokenization and Padding

```python
# Split data on token lists
train_tokens, val_tokens, train_labels, val_labels = train_test_split(
    df_copy['tokens'], df_copy['label'], test_size=0.2, random_state=42)

# Fit tokenizer on train token lists
max_vocab = 10000
tokenizer = Tokenizer(num_words=max_vocab, oov_token='<OOV>')
# texts here can be lists of tokens
tokenizer.fit_on_texts(train_tokens)

# Convert token lists to sequences
max_len = 100
train_seq = pad_sequences(
    tokenizer.texts_to_sequences(train_tokens), maxlen=max_len, padding='post'
)
val_seq   = pad_sequences(
    tokenizer.texts_to_sequences(val_tokens),   maxlen=max_len, padding='post'
)

vocab_size = min(max_vocab, len(tokenizer.word_index) + 1)

# Rebuild embedding matrix for tokens
embedding_matrix = np.zeros((vocab_size, embedding_dim), dtype='float32')
for word, i in tokenizer.word_index.items():
    if i >= vocab_size: continue
    vec = emb_index.get(word)
    if vec is not None:
        embedding_matrix[i] = vec
```

## Prepare tf.data Datasets

```python
train_ds = tf.data.Dataset.from_tensor_slices((train_seq, train_labels))
train_ds = train_ds.shuffle(buffer_size=len(train_seq), seed=42).batch(32).prefetch(tf.data.AUTOTUNE)
val_ds   = tf.data.Dataset.from_tensor_slices((val_seq,   val_labels))
val_ds   = val_ds.batch(32).prefetch(tf.data.AUTOTUNE)
```

## Build, Train & Evaluate BiLSTM

```python
early_stop = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)
model_bi = Sequential([
    Embedding(input_dim=vocab_size, output_dim=embedding_dim,
              weights=[embedding_matrix], input_length=max_len, trainable=False),
    Bidirectional(LSTM(128)),
    Dropout(0.5),
    Dense(64, activation='relu'),
    Dense(df_copy['label'].nunique(), activation='softmax')
])
model_bi.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
model_bi.summary()
hist_bi = model_bi.fit(train_ds, validation_data=val_ds, epochs=20, callbacks=[early_stop])

y_true, y_pred = [], []
for seq_batch, label_batch in val_ds:
    preds = model_bi.predict(seq_batch)
    y_pred.extend(np.argmax(preds, axis=1).tolist())
    y_true.extend(label_batch.numpy().tolist())
print("\nBiLSTM Classification Report:")
print(classification_report(y_true, y_pred))
cm = confusion_matrix(y_true, y_pred)

plt.figure(figsize=(5,4))
sns.heatmap(cm, annot=True, fmt='d')
plt.title("BiLSTM Confusion Matrix"); plt.show()
```

## Build, Train & Evaluate Text‑CNN

```python
model_cnn = Sequential([
    Embedding(input_dim=vocab_size, output_dim=embedding_dim,
              weights=[embedding_matrix], input_length=max_len, trainable=False),
    Conv1D(128, 3, activation='relu'),
    GlobalMaxPooling1D(),
    Dropout(0.5),
    Dense(64, activation='relu'),
    Dense(df_copy['label'].nunique(), activation='softmax')
])
model_cnn.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
hist_cnn = model_cnn.fit(train_ds, validation_data=val_ds, epochs=20, callbacks=[early_stop])

y_true, y_pred = [], []
for seq_batch, label_batch in val_ds:
    preds = model_cnn.predict(seq_batch)
    y_pred.extend(np.argmax(preds, axis=1).tolist())
    y_true.extend(label_batch.numpy().tolist())
print("\nText-CNN Classification Report:")
print(classification_report(y_true, y_pred))
cm = confusion_matrix(y_true, y_pred)
plt.figure(figsize=(5,4)); sns.heatmap(cm, annot=True, fmt='d'); plt.title("Text-CNN Confusion Matrix"); plt.show()
```

## 5‑Fold cross‑validation for BiLSTM

```python
X_all = pad_sequences(tokenizer.texts_to_sequences(df_copy['tokens']), maxlen=max_len, padding='post')
y_all = df_copy['label'].values
kf = KFold(n_splits=5, shuffle=True, random_state=42)
cv_scores_bi = []
for fold, (train_idx, val_idx) in enumerate(kf.split(X_all)):
    X_t, X_v = X_all[train_idx], X_all[val_idx]
    y_t, y_v = y_all[train_idx], y_all[val_idx]
    ds_t = tf.data.Dataset.from_tensor_slices((X_t, y_t)).shuffle(len(train_idx), seed=fold).batch(32)
    ds_v = tf.data.Dataset.from_tensor_slices((X_v, y_v)).batch(32)
    model_cv = Sequential([
        Embedding(vocab_size, embedding_dim, weights=[embedding_matrix], trainable=False),
        Bidirectional(LSTM(128)),
        Dropout(0.5),
        Dense(64, activation='relu'),
        Dense(df_copy['label'].nunique(), activation='softmax')
    ])
    model_cv.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
    model_cv.fit(ds_t, validation_data=ds_v, epochs=20,
                 callbacks=[EarlyStopping('val_loss', patience=3, restore_best_weights=True)], verbose=0)
    _, acc = model_cv.evaluate(ds_v, verbose=0)
    cv_scores_bi.append(acc)
print(f"BiLSTM 5-Fold CV accuracies: {cv_scores_bi}")
print(f"Mean BiLSTM CV accuracy: {np.mean(cv_scores_bi):.4f}")
```

## 5‑Fold cross‑validation for Text-CNN

```python
cv_scores_cnn = []
for fold, (train_idx, val_idx) in enumerate(kf.split(X_all)):
    X_t, X_v = X_all[train_idx], X_all[val_idx]
    y_t, y_v = y_all[train_idx], y_all[val_idx]
    ds_t = tf.data.Dataset.from_tensor_slices((X_t, y_t)) \
                    .shuffle(len(train_idx), seed=fold) \
                    .batch(32).prefetch(tf.data.AUTOTUNE)
    ds_v = tf.data.Dataset.from_tensor_slices((X_v, y_v)) \
                    .batch(32).prefetch(tf.data.AUTOTUNE)
    model_cv = Sequential([
        Embedding(input_dim=vocab_size, output_dim=embedding_dim,
                  weights=[embedding_matrix], trainable=False),
        Conv1D(128, 3, activation='relu'),
        GlobalMaxPooling1D(),
        Dropout(0.5),
        Dense(64, activation='relu'),
        Dense(df_copy['label'].nunique(), activation='softmax')
    ])
    model_cv.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
    model_cv.fit(ds_t, validation_data=ds_v, epochs=20,
                 callbacks=[EarlyStopping(monitor='val_loss', patience=3, restore_best_weights=True)], verbose=0)
    _, acc = model_cv.evaluate(ds_v, verbose=0)
    cv_scores_cnn.append(acc)
print(f"Text-CNN 5-Fold CV accuracies: {cv_scores_cnn}")
print(f"Mean Text-CNN CV accuracy: {np.mean(cv_scores_cnn):.4f}")
```

## Hyperparameter tuning via GridSearchCV for BiLSTM

```python
def build_bilstm_tuner(hp):
    model = Sequential()
    # Embedding layer with preloaded GloVe weights
    model.add(Embedding(
        input_dim=vocab_size,
        output_dim=embedding_dim,
        weights=[embedding_matrix],
        input_length=max_len,
        trainable=False
    ))
    # Tune the number of LSTM units
    model.add(Bidirectional(
        LSTM(units=hp.Choice('units', [32, 64, 128]))
    ))
    # Tune dropout rate
    model.add(Dropout(rate=hp.Choice('dropout', [0.3, 0.5, 0.7])))
    model.add(Dense(64, activation='relu'))
    # Output layer
    model.add(Dense(df_copy['label'].nunique(), activation='softmax'))

    model.compile(
        optimizer='adam',
        loss='sparse_categorical_crossentropy',
        metrics=['accuracy']
    )
    return model

# 2) Instantiate the RandomSearch tuner
bilstm_tuner = kt.RandomSearch(
    hypermodel=build_bilstm_tuner,
    objective='val_accuracy',
    max_trials=6,
    executions_per_trial=1,
    directory='bilstm_tuning',
    project_name='smm'
)

# 3) Run the search on NumPy arrays (not tf.data.Dataset)
#    train_seq, train_labels, val_seq, val_labels must already be defined
bilstm_tuner.search(
    x=train_seq,
    y=train_labels,
    validation_data=(val_seq, val_labels),
    epochs=20,
    callbacks=[early_stop],
    verbose=1
)

# 4) Retrieve the best hyperparameters
best_bi_hp = bilstm_tuner.get_best_hyperparameters(num_trials=1)[0]
print("Best BiLSTM hyperparameters:", best_bi_hp.values)
```

## Hyperparameter tuning via GridSearchCV for Text‑CNN
